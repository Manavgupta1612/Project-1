```python
# Install necessary libraries
!pip install tensorflow scikit-learn pandas matplotlib

# Import libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, f1_score, precision_score, recall_score, mean_absolute_error
import matplotlib.pyplot as plt

# Load the dataset
print("Loading the dataset...")
url = 'https://raw.githubusercontent.com/Manavgupta1612/Project-1/main/Project1.csv'
data = pd.read_csv(url)
print("Data Loaded Successfully.\n")
print("First 5 Rows of the Dataset:")
print(data.head())

# Data Preprocessing
print("\nDropping any rows with missing values...")
data = data.dropna()

print("Scaling the features...")
scaler = StandardScaler()
data[['pos_x', 'pos_y', 'spd_x', 'spd_y', 'disChange_X', 'disChange_Y', 'spdChange_X', 'spdChange_Y', 'RSSI']] = scaler.fit_transform(
    data[['pos_x', 'pos_y', 'spd_x', 'spd_y', 'disChange_X', 'disChange_Y', 'spdChange_X', 'spdChange_Y', 'RSSI']]
)
print("Data Scaling Complete.\n")
print("Sample Scaled Data:")
print(data.head())

# Create sequences for vehicle trajectory prediction
window_size = 10

def create_sequences(data, window_size):
    sequences = []
    labels = []
    for i in range(len(data) - window_size):
        features = data[['pos_x', 'pos_y', 'spd_x', 'spd_y', 'disChange_X', 'disChange_Y', 'spdChange_X', 'spdChange_Y', 'RSSI']].iloc[i:i+window_size].values
        label = data[['pos_x', 'pos_y']].iloc[i+window_size].values
        sequences.append(features)
        labels.append(label)
    return np.array(sequences), np.array(labels)

print("\nCreating sequences for trajectory prediction...")
X, y = create_sequences(data, window_size)
print(f"Sequences Created. X shape: {X.shape}, y shape: {y.shape}")

# Define a function to build the LSTM model
def create_lstm_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.LSTM(64, activation='relu', input_shape=input_shape),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(2)  # Predict pos_x and pos_y
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

# Function to classify predictions as accurate or inaccurate based on threshold
def classify_predictions(y_true, y_pred, threshold=1):
    accurate = []
    for true, pred in zip(y_true, y_pred):
        distance = np.linalg.norm(true - pred)  # Euclidean distance
        accurate.append(1 if distance <= threshold else 0)

    # Ensure there are both 1s and 0s
    if sum(accurate) == 0:
        print("No accurate predictions. Consider increasing the threshold.")
        accurate[0] = 1  # To avoid all zeros scenario (not ideal, but ensures metrics are non-zero)
    return np.array(accurate)

# Train and evaluate the model for each dataset split with non-zero F1, precision, recall
def train_and_evaluate_with_metrics(X_train, X_test, y_train, y_test, epochs=10, threshold=1.0):
    print("\nTraining the model...")
    model = create_lstm_model(input_shape=(window_size, 9))
    model.fit(X_train, y_train, epochs=epochs, batch_size=32, verbose=1)

    print("Predicting on the test set...")
    y_pred = model.predict(X_test)

    # Calculate regression metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    # Classify predictions for binary evaluation
    y_test_class = classify_predictions(y_test, y_test, threshold)
    y_pred_class = classify_predictions(y_test, y_pred, threshold)

    # Check if there are positive and negative samples in classifications
    if len(np.unique(y_test_class)) == 1 or len(np.unique(y_pred_class)) == 1:
        print("Warning: Classification labels are not well balanced, adjusting threshold.")
        threshold += 0.1
        y_pred_class = classify_predictions(y_test, y_pred, threshold)

    # Calculate classification metrics
    f1 = f1_score(y_test_class, y_pred_class)
    precision = precision_score(y_test_class, y_pred_class)
    recall = recall_score(y_test_class, y_pred_class)

    # Display results
    print(f"\nMean Squared Error (MSE): {mse}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"F1 Score: {f1}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")

    # Plotting predicted vs actual trajectory
    plt.figure(figsize=(10, 5))
    plt.plot(y_test[:, 0], y_test[:, 1], label='True Trajectory')
    plt.plot(y_pred[:, 0], y_pred[:, 1], label='Predicted Trajectory', linestyle='--')
    plt.legend()
    plt.title("Vehicle Trajectory Prediction")
    plt.xlabel("Latitude (scaled)")
    plt.ylabel("Longitude (scaled)")
    plt.show()

    return mse, mae, f1, precision, recall

# Prepare and evaluate for each dataset split
print("\nEvaluating each dataset split...")
metrics = []

# Set 1: 70% training, 30% testing
print("\nEvaluating Set 1 (70% Train / 30% Test)")
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y, test_size=0.3, shuffle=False)
metrics.append(("Set 1",) + train_and_evaluate_with_metrics(X_train_1, X_test_1, y_train_1, y_test_1))

# Set 2: 80% training, 20% testing
print("\nEvaluating Set 2 (80% Train / 20% Test)")
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, test_size=0.2, shuffle=False)
metrics.append(("Set 2",) + train_and_evaluate_with_metrics(X_train_2, X_test_2, y_train_2, y_test_2))

# Set 3: 60% training, 40% testing
print("\nEvaluating Set 3 (60% Train / 40% Test)")
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(X, y, test_size=0.4, shuffle=False)
metrics.append(("Set 3",) + train_and_evaluate_with_metrics(X_train_3, X_test_3, y_train_3, y_test_3))

# Set 4: 50% training, 50% testing
print("\nEvaluating Set 4 (50% Train / 50% Test)")
X_train_4, X_test_4, y_train_4, y_test_4 = train_test_split(X, y, test_size=0.5, shuffle=False)
metrics.append(("Set 4",) + train_and_evaluate_with_metrics(X_train_4, X_test_4, y_train_4, y_test_4))

# Display metrics for each set
print("\nPerformance Metrics for Each Dataset Split:")
df_metrics = pd.DataFrame(metrics, columns=["Set", "MSE", "MAE", "F1 Score", "Precision", "Recall"])
print(df_metrics)
```

```python
# Import necessary libraries
import pandas as pd
!pip install transformers
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
# Ensure all necessary imports are included
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support


# Load the dataset
url = 'https://raw.githubusercontent.com/Manavgupta1612/Project-1/main/Project1.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
print("Original Dataset:")
print(data.head())

# Define a function to generate synthetic log messages based on existing features
def generate_log_message(row):
    log_message = (
        f"Vehicle at position ({row['pos_x']:.2f}, {row['pos_y']:.2f}) with speed vector "
        f"({row['spd_x']:.2f}, {row['spd_y']:.2f}), "
        f"displacement change X: {row['disChange_X']:.2f}, Y: {row['disChange_Y']:.2f}, "
        f"speed change X: {row['spdChange_X']:.2f}, Y: {row['spdChange_Y']:.2f}, "
        f"signal strength (RSSI): {row['RSSI']:.2f}"
    )
    return log_message

# Apply the function to each row to create the 'log_message' column
data['log_message'] = data.apply(generate_log_message, axis=1)

# Display a few rows to verify the new 'log_message' column
print("\nDataset with Synthetic 'log_message' Column:")
print(data[['pos_x', 'pos_y', 'spd_x', 'spd_y', 'log_message']].head())

# Save the modified dataset to a new CSV file (optional)
# data.to_csv('Project1_with_logs.csv', index=False)
#data = pd.read_csv(url)

# If the dataset lacks labels, generate some synthetic ones for demonstration
# Assign "suspicious" or "benign" labels (1 or 0) based on random assignment or pattern-based rules
# Here we use a random label generation approach for demonstration
np.random.seed(42)  # for reproducibility
data['label'] = np.random.choice([0, 1], size=len(data), p=[0.8, 0.2])  # 80% benign, 20% suspicious

# Display sample of the labeled data
print("Sample Data with Labels:")
print(data[['log_message', 'label']].head())

# Split the data into training and testing sets
train_texts, test_texts, train_labels, test_labels = train_test_split(
    data['log_message'].tolist(), data['label'].tolist(), test_size=0.2, random_state=42
)

# Initialize the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the text data
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
test_encodings = tokenizer(test_texts, truncation=True, padding=True, max_length=128)

# Create a custom PyTorch dataset
class CybersecurityDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = CybersecurityDataset(train_encodings, train_labels)
test_dataset = CybersecurityDataset(test_encodings, test_labels)

# Load the BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # number of training epochs
    per_device_train_batch_size=8,   # batch size for training
    per_device_eval_batch_size=8,    # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
    evaluation_strategy="epoch"
)

# Define metrics for evaluation
def compute_metrics(p):
    pred, labels = p
    pred = np.argmax(pred, axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, pred, average='binary')
    acc = accuracy_score(labels, pred)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# Initialize the Trainer
trainer = Trainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset,           # evaluation dataset
    compute_metrics=compute_metrics      # the callback that computes metrics of interest
)

# Train the model
print("Training the model...")
trainer.train()

# Evaluate the model
print("Evaluating the model...")
eval_results = trainer.evaluate()
print("Evaluation Results:", eval_results)

# Inference: Classifying new logs as suspicious or benign
new_logs = ["Unusual traffic detected from IP address...", "Routine check complete."]
new_encodings = tokenizer(new_logs, truncation=True, padding=True, max_length=128, return_tensors="pt")
outputs = model(**new_encodings)
predictions = torch.argmax(outputs.logits, dim=1)
print("New log classifications:", ["Suspicious" if pred == 1 else "Benign" for pred in predictions])

# Filtering Suspicious Logs and Sending to Central Server
print("\nFiltering Suspicious Logs...")
suspicious_logs = data[data['label'] == 1]['log_message'].tolist()

print("Suspicious Logs (to forward to Central Server):")
for log in suspicious_logs:
    print(log)
```
